```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,100,1000)
y = np.sin(x)
plt.plot(x,y)
```

```python
fig, ax = plt.subplots(2,1,figsize=(12,5),sharex=True)
ax[0].plot(x,y,'r--',lw=3)
ax[1].plot(x,np.cos(x),'b*',lw=2)
ax[0].set_xlim(0,80)
ax[0].set_ylim(-2,2)
ax[0].set_title('this is Y=sin(x)')
ax[1].set_xlabel('x')
ax[0].set_ylabel('Y=sin(x)')
```
