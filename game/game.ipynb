```python
import numpy as np
import pandas as pd
```

```python
def get_userinput():
    usercode = input('my choice is: ')
    usercode = usercode.lower()
    if usercode in ['scissor','paper','stone']:
        return(usercode)
    else:
        print('error input')
        return('error')
def get_compintput():
    compinput = np.random.randint(0,2)
    if compinput == 0:
        return('scissor')
    elif compinput == 1:
        return('paper')
    else:
        return('stone')
```

```python
userinput = get_userinput()
compintput = get_compintput()
print('my input is :' + userinput,'computer input is :' + compintput)
if userinput == compintput:
    print('no one win')
elif userinput == 'scissor':
    if compintput == 'paper':
        print('user wins')
    else:
        print('comp wins')
elif userinput == 'paper':
    if compintput == 'stone':
        print('user wins')
    else:
        print('comp wins')
elif userinput == 'stone':
    if compintput == 'scissor':
        print('user wins')
    else:
        print('comp wins')
else:
    print('not a complete game')
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-2,2,1500)
y1 = np.sqrt(1-(abs(x)-1)**2)
y2 = -3*np.sqrt(1-(abs(x)/2)**0.5)
```

```python
fig, axes = plt.subplots(2,1,figsize = (12,10),sharex=True)
axes[0].plot(x,y1,lw=2,linestyle='--',marker='s',markeredgecolor='r')
axes[1].plot(x,y2,color='red')
plt.xlabel('xx')
plt.ylabel('yy')
#axes.get_xlabel('vlaue x',fontsize=20)
#set_shared_ylabel(axes,'value of y1')
```
